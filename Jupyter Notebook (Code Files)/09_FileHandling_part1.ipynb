### Use of context manager

```python
with open(filename, 'r') as f:   # notice the difference here
    
    for line in f:
        print(line.strip())


        
# also, no close call needed now ... so you can't forget it!
```

```python
with open(filename, 'r') as f:
    for line in f:
        print(line)
        print(f.read(1))
        
        print(f.read(1))
    
    # the whole file
```

```python
with open(filename, 'r') as f:
    print(f.read(1000))
```

```python
with open(out_filename, 'a') as f:   # Let's call this a couple of times
    f.write("Two...")
```

```python
with open(out_filename, 'a') as f:   # Let's call this a couple of times
    f.write("Three...\n")
```

```python
with open(out_filename, 'a') as f:   # Let's call this a couple of times
    f.write("One...")
    f.read()
    print(f.tell())
```

```python
with open(out_filename, 'r+') as f:   # Let's call this a couple of times
    f.seek(4)
    print(f.read())
    f.write('a')
    print(f.read())
```
